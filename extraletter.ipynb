You are given two strings s and t.
String t is generated by random shuffling string s and then adding one more letter at a random position.
Return the letter that was added to t.
Example 1:
Input: s = "abcd", t = "abcde"
Output: "e"
Explanation: 'e' is the letter that was added.
Example 2:
Input: s = "", t = "y"
Output: "y"
 Example 3:
Input: s = "weght", t = "gexhtw"
Output: "x"

```python
def extraLetter(s1, s2):
    # return set(s2) - set(s1)
    letters1 = {x : s1.count(x) for x in s1}
    letters2 = {x : s2.count(x) for x in s2}
    for letter in letters2.keys():
        if letter not in letters1.keys():
            return letter 
        elif letters2[letter] > letters1[letter]:
            return letter
```

```python
s = "abcd" 
t = "abcde"
print(extraLetter(s, t))
```

```python
s = ""
t = "y"
print(extraLetter(s, t))
```

```python
s = "weght"
t = "gexhtw"
print(extraLetter(s, t))
```

```python
s = "abcd"
t = "abadc"
print(extraLetter(s, t))
```

```python
def addedLetter(s1, s2):
    sumS1 = 0
    sumS2 = 0
    for letter in s1:
        sumS1 += ord(letter)
    for letter in s2: 
        sumS2 += ord(letter)
    return chr(sumS2 - sumS1)
```

```python
s = "abcd"
t = "abadc"
print(addedLetter(s, t))
```
